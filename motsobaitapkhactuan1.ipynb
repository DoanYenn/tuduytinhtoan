Một số bài tập khác tuần 1

BÀI 1: Kiểm tra số chẵn lẻ: Nhập một số nguyên, kiểm tra và in ra số đó là số chẵn hay số lẻ.

```python
n = int(input("Nhập một số nguyên: "))
if n % 2 == 0:
    print(f"{n} là số chẵn")
else:
    print(f"{n} là số lẻ")
```

BÀi 2: Kiểm tra số có tận cùng là 5: Nhập một số nguyên, nếu tận cùng là 5 in ra "True", ngược lại in ra "False".

```python
n = int(input("Nhập một số nguyên: "))
if n % 10 == 5:
    print("True")
else:
    print("False")
```

BÀI 3: Kiểm tra chia hết cho 3 và 5: Nhập một số nguyên, nếu chia hết cho cả 3 và 5 in ra "True", ngược lại in ra "False".

```python
n = int(input("Nhập một số nguyên: "))
if n % 3 == 0 and n % 5 == 0:
    print("True")
else:
    print("False")
```

BÀI 4: Kiểm tra tuổi bầu cử: Nhập năm sinh, kiểm tra người đó có đủ 18 tuổi để đi bầu cử hay không.

```python
nam_sinh = int(input("Nhập năm sinh của bạn: "))
tuoi = 2026 - nam_sinh
if tuoi >= 18:
    print("Đủ điều kiện bầu cử")
else:
    print("Chưa đủ điều kiện bầu cử")
```

BÀI 5: So sánh 2 số: Nhập vào hai số, in ra số lớn hơn hoặc thông báo nếu hai số bằng nhau.

```python
a = float(input("Nhập số thứ nhất: "))
b = float(input("Nhập số thứ hai: "))
if a > b:
    print(f"Số lớn hơn là: {a}")
elif b > a:
    print(f"Số lớn hơn là: {b}")
else:
    print("Hai số bằng nhau")
```

BÀI 6: Kiểm tra ký tự là chữ hay số: Nhập một ký tự, kiểm tra xem ký tự đó là chữ cái hay chữ số.

```python
ky_tu = input("Nhập một ký tự: ")
if ky_tu.isdigit():
    print(f"'{ky_tu}' là chữ số")
elif ky_tu.isalpha():
    print(f"'{ky_tu}' là chữ cái")
else:
    print(f"'{ky_tu}' là ký tự đặc biệt")
```

BÀI 7: Tính điểm học lực: Nhập điểm (0-10), phân loại: >=8: Giỏi; >=6.5 và <8: Khá; >=5 và <6.5: Trung bình; <5: Yếu.

```python
diem = float(input("Nhập điểm (0-10): "))
if diem >= 8:
    print("Giỏi")
elif diem >= 6.5:
    print("Khá")
elif diem >= 5:
    print("Trung bình")
else:
    print("Yếu")
```

BÀI 8: Kiểm tra năm nhuận: Kiểm tra năm đó có phải năm nhuận không (Chia hết cho 400 HOẶC chia hết cho 4 nhưng không chia hết cho 100).

```python
nam = int(input("Nhập năm: "))
if (nam % 400 == 0) or (nam % 4 == 0 and nam % 100 != 0):
    print(f"{nam} là năm nhuận")
else:
    print(f"{nam} không phải năm nhuận")
```

BÀI 9: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng
(0 → "Không", 1 → "Một", …).

```python
n = int(input("Nhập số từ 0 đến 9: "))

match n:
    case 0:
        print("Không")
    case 1:
        print("Một")
    case 2:
        print("Hai")
    case 3:
        print("Ba")
    case 4:
        print("Bốn")
    case 5:
        print("Năm")
    case 6:
        print("Sáu")
    case 7:
        print("Bảy")
    case 8:
        print("Tám")
    case 9:
        print("Chín")
    case _:
        print("Số không hợp lệ")
```

BÀI 10: Kiểm tra điểm thi xem có qua môn hay phải học lại:
Nhập điểm thi, nếu điểm ≥ 4 thì in "Qua môn", ngược lại in "Học lại".

```python
diem = float(input("Nhập điểm thi: "))

if diem >= 4:
    print("Qua môn")
else:
    print("Học lại")
```

BÀI 11: 11. Kiểm tra năm sinh và tính tuổi:
Nhập năm sinh, tính tuổi hiện tại (giả sử năm hiện tại là 2025), và kiểm tra xem người đó có đủ 18 tuổi hay chưa.

```python
nam_sinh = int(input("Nhập năm sinh: "))

tuoi = 2025 - nam_sinh

print("Tuổi:", tuoi)

if tuoi >= 18:
    print("Đủ 18 tuổi")
else:
    print("Chưa đủ 18 tuổi")
```
