BAI 3: Nhap vao a, b, c bat ki. kiem tra xem day co phai la 3 canh cua mot tam giac khong, nau khong in ra man hinh 'Khong phai la mot tam giac'. nguoc lai kiem tra la tam giac gi. Tinh dien tich chu vi cua tam giac va in ra man hinh 

```python
import math 
a, b, c = map(float, input().split())
if a + b > c and a + c > b and b + c > a : 
    C = a + b + c
    p = C / 2
    S = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Chu vi tam giac {C}")
    print(f"Dien tich tam giac {S}")
    if a == b and b == c :
        print("Tam giac deu")
    elif a == b or a == c or b == c :
        print("Tam giac can")
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        print("Tam giac vuong")
    else :
        print("Tam giac thuong")
else:
    print("Khong phai tam giac")                      
```

BAI 5: Viet chuong trinh giai phuong trinh bac 2

```python
import math 
a, b, c = map(float, input().split())
delta = b * b - (4 * a * c)
if a == 0:
    print("khong la phuong trinh bac 2")
else :    
    if delta < 0 :
       print("Phuong trinh vo nghiem ")
    elif delta == 0 :
       print("phuong trinh co nghiem kep")
       x = -b / 2 * a
       print(f"nghiem cua phuong trinh : {x}")
    else :
       print("phuong trinh co 2 nghiem phan biet")
       x1 = (-b + math.sqrt(delta)) / (2 * a) 
       x2 = (-b - math.sqrt(delta)) / (2 * a)    
       print(f"nghiem thu nhat: {x1}")
       print(f"nghiem thu 2: {x2}")   
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?
