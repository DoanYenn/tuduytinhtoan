### W4A1: Viết chương trình python sử dụng 1 trong 2 vòng lặp for hoặc while để tính tổng các số nguyên từ 1 đến n. (n <= 1000)

```python
#for
n = int(input())
tong = 0
for i in range(1, n + 1):
    tong += i 
print(f"{tong}")  
```

```python
#while
n = int(input())
tong = 0
i = 1
while i <= n:
    tong += i
    i += 1
print(f"{tong}")     
```

### W4A7: Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n.

```python
n = int(input())
max_nt = 0
if n < 2:
    print("nhap so khac")
else:
    for i in range(2, n + 1):
        if n % i == 0:
            nt = True
            for j in range(2, i):
                if i % j == 0:
                   nt = False
            if nt:
                max_nt = i
print(max_nt)         
    


```

### W4A8:Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

```python
n = int(input())
step = 0
while True:
    temp = n
    rev = 0
    while temp > 0:
        rev = rev * 10 + temp % 10
        temp //= 10
    if n == rev:
        break
    n = n + rev
    step += 1
print(step, n)        
```
