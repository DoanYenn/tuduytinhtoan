BÀi 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ta xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.

```python
a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))
s_hcn = a * b
r = a / 2
s_tron = 3.14 * (r ** 2)
s_con_lai = s_hcn - s_tron
print(f"{s_con_lai:.2f}")
```

BÀI 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó.

```python
c = input("Nhập kí tự: ")

if c.isupper():
    print(c.lower())
elif c.islower():
    print(c.upper())
else:
    print("Không phải chữ cái")
```

BÀI 4: Viết chương trình nhập vào một “kí tự”, kiểm tra xem c có phải là kí tự chữ cái không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tự alphabet”, nếu sai in ra “{c} không phải là kí tự alphabet”.

```python
c = input("Nhập kí tự: ")

if c.isalpha():
    print(f"{c} là kí tự alphabet")
else:
    print(f"{c} không phải là kí tự alphabet")
```

BÀI 6: Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
import math

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    S = math.sqrt(p * (p-a) * (p-b) * (p-c))
    print(round(S,1))
else:
    print("Khong phai 3 canh cua tam giac") 
```

Từ 0 – 50 kWh: 1.984 đồng/kWh

Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh

Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh

Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh

Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh

Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh

```python
ten = input("Ten chu ho: ")
truoc = int(input("Chi so thang truoc: "))
nay = int(input("Chi so thang nay: "))

kwh = nay - truoc
tien = 0

if kwh <= 50:
    tien = kwh * 1984

elif kwh <= 100:
    tien = 50*1984 + (kwh-50)*2050

elif kwh <= 200:
    tien = 50*1984 + 50*2050 + (kwh-100)*2380

elif kwh <= 300:
    tien = 50*1984 + 50*2050 + 100*2380 + (kwh-200)*2998

elif kwh <= 400:
    tien = 50*1984 + 50*2050 + 100*2380 + 100*2998 + (kwh-300)*3350

else:
    tien = 50*1984 + 50*2050 + 100*2380 + 100*2998 + 100*3350 + (kwh-400)*3460

tien = int(tien * 1.08)

print("Ho va ten:", ten)
print("Tien phai tra la:", tien)
```
