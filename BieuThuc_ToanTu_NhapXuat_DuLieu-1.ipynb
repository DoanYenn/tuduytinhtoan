W2A3: Viết chương trình nhận vào hai số nguyên a, b từ bàn phím. In ra tổng, hiệu, tích, phần nguyên, phần dư và kết quả chia thực của phép chia a cho b (lấy đến 2 chữ số phần thập phân).

W2A4: Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng 6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:Các điểm a1, b1, c1 được nhân hệ số 1.Các điểm a2, b2 được nhân hệ số 2.Điểm a3 được nhân hệ số 3.Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)*2 + a3*3)/10. Lấy đến 1 chũ số phần thập phân

```python
a1, b1, c1, a2, b2, a3 = map(float, input().split())
tong_diem = (a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3
diem_tb = tong_diem / 10
print(f"Diem Trung binh: {diem_tb:.1f}")
```
