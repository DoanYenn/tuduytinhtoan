### Bài 4: in ra màn hình tam giác cân có độ cao h nhập từ bàn phím

```python
h = int(input("Nhập chiều cao h: "))
for i in range(1, h + 1):
    if i == 1:
        print(" " * (h - i) + "*")
    elif i == h:
        print("*" * (2 * h - 1))
    else:
        print(" " * (h - i) + "*" + " " * (2 * (i - 1) - 1) + "*")
```

### Bài 12: Nhập số nguyên dương n. Cho S(k) = 1 + 2 + 3 + ... + k
### Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input())
S = 0
k = 0
while S + (k + 1) < n:
    k += 1
    S += k
print(f"k= {k}")    


```

### Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho: 1 + 1/2 + 1/3 + ... +1/n > A

```python
A = int(input())
tong = 0
n = 0
while tong <= A:
    n += 1
    tong += 1/n
print(f"n = {n}")    
```

### Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
so_dau = int(input())
max_day = so_dau
min_day = so_dau
while True:
    x = int(input())
    if  x == -1:
        break
    if x > max_day:
       max_day = x
    if x < min_day:
       min_day = x    
print(f"so lon nhat: {max_day}")
print(f"so nho nhat: {min_day}")
```

### Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ.

```python
n = int(input())
dem_chan = 0
dem_le = 0
while n > 0:
    chu_so = n % 10
    if chu_so % 2 == 0:
        dem_chan += 1
    else:
        dem_le += 1
    n //= 10    
print(f"chu so chan: {dem_chan}") 
print(f"chu so le: {dem_le}")               
```
